Resumo estatístico e analítico da pesquisa própria levantada para mensurar e compreender o mercado e os potenciais clientes.

### Preparação do Ambiente

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
sheet_id = "1gd2aScU5EdnkaQH2FPS6d2XRcgmTuUg2mVeHAuPgGl4"
sheet_name = "Pesquisa"
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"
```

### Análises Gerais

```python
df = pd.read_csv(url)
df.head()
```

```python
df.describe()
```

```python
counts = df['Já utilizou algum sistema de monitoramento antes?'].value_counts()

# Plot
plt.figure(figsize=(8, 6))
plt.rcParams['font.family'] = 'Roboto'  # Set font family to Times
colors = ['#66c2a5', '#fc8d62']  # Custom colors
explode = (0.05, 0)  # Explode the first slice slightly
counts.plot(kind='pie', autopct='%1.1f%%', colors=colors, explode=explode, startangle=120)
plt.title('Já utilizou algum sistema de monitoramento?', fontname='Roboto', fontsize=14)
plt.ylabel('')  # Remove y-label
plt.xlabel('')  # Remove y-label
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add legend
plt.legend(labels=counts.index, loc='upper right', bbox_to_anchor=(1.1, 0.8))
```

```python
counts = df['Qual dos seguintes aspectos considera mais importante monitorar?'].value_counts()

# Plot
plt.figure(figsize=(8, 6))
plt.rcParams['font.family'] = 'Roboto'  # Set font family to Times
colors = ['#66c2a5', '#fc8d62', '#fcfd63']  # Custom colors
explode = (0.05, 0.05, 0.05)  # Explode the first slice slightly
counts.plot(kind='pie', autopct='%1.1f%%', colors=colors, explode=explode, startangle=120)
plt.title('O mais importante para se monitotar', fontname='Roboto', fontsize=14)
plt.ylabel('')  # Remove y-label
plt.xlabel('')  # Remove y-label
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add legend
plt.legend(labels=counts.index, loc='upper right', bbox_to_anchor=(1.1, 0.8))
```

```python
distribution_data = df['Quanto pagaria para obter um sistema de monitoramento? (em R$)'].value_counts()

# Plot
plt.figure(figsize=(8, 6))
plt.rcParams['font.family'] = 'Roboto'  # Set font family to Times
distribution_data.plot(kind='bar', color='#66c2a5')
plt.title('Distribuição de Preços', fontname='Roboto', fontsize=14)
plt.xlabel('Categories', fontname='Roboto', fontsize=12)
plt.ylabel('Frequency', fontname='Roboto', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate x-labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines for better visualization

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
def heightChecker(heights):
        ordered_heights = sorted(heights)
        non_position = 0
        for i in range(len(heights)):
            if ordered_heights[i] != heights[i]:
                non_position += 1
        return non_position
```

```python
heightChecker([1,2,3,4,5])
```
